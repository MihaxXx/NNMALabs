```python
import torch
```

```python
import torchvision
```

```python
from torchvision.models import resnet152
```

```python
import torchvision.transforms as T
```

```python
model=resnet152(True)
```

```python
model.eval()
```

```python
impt=torch.rand(1,3,224,224)
```

```python
out = model(impt).detach()
```

```python
class_idx = out[0].argmax()
```

```python
print(class_idx)
```

```python
with open('imagenet-classes.txt','r') as f:
    classes_names = [
        line.rstrip()
        for line in f
    ]
```

```python
print(classes_names[class_idx])
```

```python
from PIL import Image
```

```python
img = Image.open('/Users/student/Documents/1200x900.webp').convert('RGB')
```

```python
trans = T.Compose([T.Resize((224,224)),T.ToTensor()])
```

```python
timg = trans(img).unsqueeze(0)
```

```python
%%time
out = model(timg).detach()
```

```python
pred = out[0].argmax()
```

```python
print(classes_names[pred])
```

```python
rten = torch.rand(1,3,225,225, requires_grad=True)
```

```python
traced_model = torch.jit.trace(model,rten)
```

```python
traced_model.save('model.pt')
```

```python
load_model = torch.jit.load('model.pt')
```

```python
load_model.eval()
```

```python
out1 = model(timg).detach()
```

```python
out2 = load_model(timg).detach()
```

```python
print(torch.linalg.norm(out1-out2))
```
